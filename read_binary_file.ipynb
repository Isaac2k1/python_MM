```python
filename = 'l18000105-0001.rba'
hugo = ''
with open(filename, "rb") as f:
    while (hugo != f.read(1)):
        hugo = f.read(1)
        while hugo:
            # Do stuff with byte.
            hugo = f.read(1)
            print(hugo)
```
